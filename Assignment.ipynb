```python
#pip install matplotlib

import matplotlib.pyplot as plt
import numpy as np


#Finding interest rate, how will this charge us

def prefund(money: float):
    return money - (0.065*money)

#monthly repayments, how much money will be paid

def monthlyRepayment(money: float):
    interest = list([100000])
    month = list([0])
    k = 0
    while True:
        balance =interest[-1] - prefund(money)
        interest.append(balance)
        k+=1
        month.append(k)
        if balance <=0:
            return month,interest
        pass
#displaying monthly repayment
x_1000,y_1000 = monthlyRepayment(1000)
x_2000,y_2000 = monthlyRepayment(2000)
x_3000,y_3000 = monthlyRepayment(3000)
x_4000,y_4000 = monthlyRepayment(4000)
    
    
    
    
plt.plot(x_1000,y_1000,label='1000')
plt.plot(x_2000,y_2000,label='2000')
plt.plot(x_3000,y_3000,label='3000')
plt.plot(x_4000,y_4000,label='4000')

#exploring our limits and label graph
plt.xlim(0,60)
plt.ylim(0,100000)
plt.legend()
plt.ylabel('Debts')
plt.xlabel('number of months')
plt.show()
```
